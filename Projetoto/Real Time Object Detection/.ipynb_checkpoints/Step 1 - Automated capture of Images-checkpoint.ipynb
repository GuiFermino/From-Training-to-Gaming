# Purpose of this code:
- This code was adapted to capture images from the webcam, preparing the environment for obtaining and categorizing the data (which in this case are diferent hand signs) that we shall label, train and finally use the trained model for real-time recognition later in this project.

```python
# Dependecies which makes possible the autmated process for data collection:

import cv2 #openCV (main depency)
#https://docs.opencv.org/master/d0/de3/tutorial_py_intro.html

import os #save image
#https://docs.python.org/3/library/os.html

import time #give a break in between colection of real time data
#https://docs.python.org/3/library/time.html

import uuid #name image files
#https://docs.python.org/3/library/uuid.html
```

```python
# Creating a path for the folder of images collection:
IMAGES_PATH = 'Tensorflow/workspace/images/collected_images'
```

```python
#labelling of poses collected:
labels = ['right','left','up','down','jump','shoot']

#Numver  of images collected:
number_of_imgs = 15 #13 for training and 2 for testing
#Here I opted for 15 images for each category (total of 90 images in this case),
#checking how my laptop would process such amount of data
```

```python
#Little code to discover which is our device (webcam) recogniton number
for x in range(25): # Let's test to catch which device  is our webcam setted in
    cap = cv2.VideoCapture(x - 2) # Capture video from webcam, number may vary from PC to PC
    print (cap.isOpened(), x - 2) # Showing what each periferic return
```

```python
#For my PC, Webcam was in 0 (could be  1 also, but it was discarted by testing in practice)
for label in labels:
    os.makedirs('Tensorflow/workspace/images/collected_images/\\'+label) #Create a folder for each label
    
    webcamTry = 0 #(0 or 1)
    
    cap = cv2.VideoCapture(webcamTry) # Capture video from webcam
    print('Collecting images for {}'.format(label)) #Warning for what category of image will be captured
    time.sleep(5) #Give time to get into position to specific category (5 seconds in this case)
    for img_num in range(number_of_imgs): #Loop for capture with the size of "number_of_imgs" 
        ret, frame = cap.read() #image of the video frame is returned here.
        img_name = os.path.join(IMAGES_PATH, label, label+'.'+'{}.jpg'.format(str(uuid.uuid1()))) #Defining image name
        # uuid helps us at not duplicate the name of the image
        cv2.imwrite(img_name, frame) #write to directory
        cv2.imshow('frame', frame) #show image in the screen
        print('Prepare for image number {}'.format(img_num))#Warning for image number which will be captured
        time.sleep(3) #Give time to get into position to get each images (3 seconds in this case)
        
        if cv2.waitKey(1) & 0xFF == ord('q'): #break key to leave loop
            break
            
    cap.release() #release a video capture
cv2.destroyAllWindows() #close windows of video capture
```

## Remarks:

    There are some practical improvments which can be made for the image captures, facilitating the training and modeling later on:
    
- The use of a clean background, even better if it has a lot of contrast like a green screen (I did the best I could in my room to get to that)
- The capture solely of the hands image, without other objects appearing in the background like a face or a t-shirt, which can "confuse" what is specifcly "hand" while training and testing the images, due to the confusing contrast (I did not think about this at the time, as you can see pictures of my hand signs in front of my face)
- The use of really diferent hand signs, in the way of not "confusing" the categorization while training and testing the images (I did not think about this at the time, as you can see my right and left hand signs categories being close to one another)
- The use of a cameraa with better resolution (I only had the implemented webcam from my computer, which is not so great...)
